```python
import pytesseract
from pdf2image import convert_from_path
from pytesseract import image_to_string
from google.colab import files
import re

uploaded = files.upload()
file_name = list(uploaded.keys())[0]

images = convert_from_path(file_name)
print(f"{len(images)} pages converted to images.")

ocr_text = ""
for i, image in enumerate(images):
    page_text = image_to_string(image)
    ocr_text += f"--- Page {i + 1} ---\n{page_text}\n"

print("Extracted Text (First 350 characters):\n", ocr_text[:350])

def extract_sections(text):
    sections = {
        "Abstract": re.search(r"(?i)\babstract\b(.*?)(?=\bkeywords\b|\bintroduction\b)", text, re.DOTALL),
        "Keywords": re.search(r"(?i)\bkeywords\b(.*?)(?=\bdate of publication\b|\babstract\b)", text, re.DOTALL),
        "Date of Publication": re.search(r"(?i)\bdate of publication\b(.*?)(?=\bjournal name\b|\bconference name\b)", text, re.DOTALL),
        "Journal Name": re.search(r"(?i)\bjournal name\b(.*?)(?=\bconference name\b|\bdoi\b)", text, re.DOTALL),
        "Conference Name": re.search(r"(?i)\bconference name\b(.*?)(?=\bdoi\b|\bvolume\b)", text, re.DOTALL),
        "DOI": re.search(r"(?i)\bdoi\b(.*?)(?=\bvolume\b|\bpage numbers\b)", text, re.DOTALL),
        "Volume & Issue Number": re.search(r"(?i)\bvolume & issue number\b(.*?)(?=\bpage numbers\b|\bintroduction\b)", text, re.DOTALL),
        "Page Numbers": re.search(r"(?i)\bpage numbers\b(.*?)(?=\bintroduction\b)", text, re.DOTALL),
        "Introduction": re.search(r"(?i)\bintroduction\b(.*?)(?=\bliterature review\b|\bmethodology\b|\bresults\b)", text, re.DOTALL),
        "Literature Review": re.search(r"(?i)\bliterature review\b(.*?)(?=\bmethodology\b|\bresults\b)", text, re.DOTALL),
        "Methodology": re.search(r"(?i)\bmethodology\b(.*?)(?=\bresults\b|\bdiscussion\b)", text, re.DOTALL),
        "Results": re.search(r"(?i)\bresults\b(.*?)(?=\bdiscussion\b|\bconclusion\b)", text, re.DOTALL),
        "Discussion": re.search(r"(?i)\bdiscussion\b(.*?)(?=\bconclusion\b)", text, re.DOTALL),
        "Conclusion": re.search(r"(?i)\bconclusion\b(.*?)(?=\breferences\b)", text, re.DOTALL),
        "References": re.search(r"(?i)\breferences\b(.*)", text, re.DOTALL),
        "Acknowledgments": re.search(r"(?i)\backnowledgments\b(.*?)(?=\breferences\b)", text, re.DOTALL),
    }

    extracted = {section: match.group(1).strip() for section, match in sections.items() if match}
    return extracted

available_sections = extract_sections(ocr_text)
print("\nAvailable Sections in the Research Paper:")
print(", ".join(available_sections.keys()) if available_sections else "No recognizable sections found.")

print("\nSelect sections to display (comma separated or type 'all' for all sections):")
print("Options: " + ", ".join(available_sections.keys()))
user_input = input("Enter your choice: ").strip().lower()
selected_sections = [s.strip().capitalize() for s in user_input.split(',')]

if "all" in user_input:
    selected_sections = list(available_sections.keys())

char_limit_options = {
    1: 500,
    2: 1000,
    3: 2000,
    4: 'all'
}

print("Select the character limit for displaying sections:")
print("1: 500 characters\n2: 1000 characters\n3: 2000 characters\n4: All characters")
option = int(input("Enter your choice (1/2/3/4): "))
char_limit = char_limit_options.get(option, 500)

def display_sections(sections, selected_sections, char_limit):
    for section in selected_sections:
        if section in sections:
            content = sections[section]
            display_content = content if char_limit == 'all' else content[:char_limit]
            print(f"\n{section}:\n{'-'*len(section)}\n{display_content}...\n")

print("\nExtracted Sections:")
display_sections(available_sections, selected_sections, char_limit)
```
